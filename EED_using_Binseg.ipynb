```python
!pip install ruptures
import numpy as np
import matplotlib.pyplot as plt
from ruptures.detection import Binseg
from ruptures.datasets import pw_constant

# Generate example data with multiple segments
n_samples = 200
n_bkps = 3

signal, _ = pw_constant(n_samples, n_features=1, n_bkps=n_bkps)

# Apply the Binseg algorithm for change point detection
model = Binseg(model="l2").fit(signal)
result = model.predict(n_bkps=n_bkps)

# Visualize the results
plt.figure(figsize=(10, 6))
plt.plot(signal, color='blue', label='Original Signal')

for bkpt in result:
    plt.axvline(x=bkpt, color='red', linestyle='--', linewidth=1, label='Change Point')

plt.title("Change Point Detection using Binseg")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.tight_layout()
plt.show()
```
